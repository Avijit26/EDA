```python
#Plot CDF 

counts, bin_edges = np.histogram(haberman_A['30'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)


counts, bin_edges = np.histogram(haberman_A['30'], bins=20, 
                                 density = True)
pdf = counts/(sum(counts))
plt.plot(bin_edges[1:],pdf);

plt.show();
```

```python
#Mean, Variance, Std-deviation,  
print("Means:")
print(np.mean(haberman_A["30"]))
#Mean with an outlier.
print(np.mean(np.append(haberman_A["30"],50)));
print(np.mean(haberman_B["30"]))


print("\nStd-dev:");
print(np.std(haberman_A["30"]))
print(np.std(haberman_B["30"]))
```

```python
#Median, Quantiles, Percentiles, IQR.
print("\nMedians:")
print(np.median(haberman_A["30"]))
#Median with an outlier
print(np.median(np.append(haberman_A["30"],50)));
print(np.median(haberman_B["30"]))



print("\nQuantiles:")
print(np.percentile(haberman_A["30"],np.arange(0, 100, 25)))
print(np.percentile(haberman_B["30"],np.arange(0, 100, 25)))


print("\n90th Percentiles:")
print(np.percentile(haberman_A["30"],90))
print(np.percentile(haberman_B["30"],90))


from statsmodels import robust
print ("\nMedian Absolute Deviation")
print(robust.mad(haberman_A["30"]))
print(robust.mad(haberman_A["30"]))

```

```python
#Box-plot can be visualized as a PDF on the side-ways.

sns.boxplot(x='1.1',y='30', data=haberman)
plt.show()
```

```python
# A violin plot combines the benefits of the previous two plots 
#and simplifies them

# Denser regions of the data are fatter, and sparser ones thinner 
#in a violin plot

sns.violinplot(x='1.1', y='30', data=haberman, size=8)
plt.show()
```

```python
#2D Density plot, contors-plot
sns.jointplot(x='30', y='64', data=haberman_A, kind="kde");
plt.show();
```
